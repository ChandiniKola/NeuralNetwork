##Naivebase Linear SVM

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score



data = pd.read_csv("glass.csv")

features = data.drop("Type", axis=1)

targets = data.iloc[ :, -1]

x_test, x_train, y_test, y_train = train_test_split(features, targets, test_size = 0.2)

# Implement the Naive Bayes method

Gau = GaussianNB( )
Gau.fit(x_train, y_train)
y_pred = Gau.predict(x_test)

# Evaluate the Naive Bayes method

print("Naive Bayes Results:")
print( "Accuracy Score = ", accuracy_score(y_test, y_pred))
print("Classification report:")
print( classification_report( y_test, y_pred ) )

# Implement the Linear SVM method

svc_linear = SVC(kernel='linear')
svc_linear.fit(x_train, y_train)
y_pred_svc = svc_linear.predict(x_test)

# Evaluate the Linear SVM method

print("Linear SVM Results:")
print("Accuracy Score = ", accuracy_score(y_test, y_pred_svc))
print("Classification report:")
print( classification_report(y_test, y_pred_svc))

```

From the above result Linear SVM has more Accuracy.

In general, Naive Bayes is a fast and simple algorithm that is well suited for high-dimensional datasets with many features. However, it can be less accurate than more complex algorithms, especially when the relationship between the features and the target is not well modeled by the assumptions of the Naive Bayes method.

Linear SVM is a more flexible and powerful algorithm that can handle a wider range of data distributions. It is also more computationally expensive than Naive Bayes, but can often produce more accurate results. The choice between Naive Bayes and Linear SVM will depend on the specific problem being solved, the nature of the data, and the computational resources available.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Salary_Data.csv")
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values

from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=1/3, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_Train, Y_Train)

Y_Pred = regressor.predict(X_Test)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_Test, Y_Pred)
print("Mean Squared Error: ", mse)

plt.scatter(X_Train, Y_Train, color='red')
plt.plot(X_Train, regressor.predict(X_Train), color='blue')
plt.title('Salary vs Experience (Train set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
